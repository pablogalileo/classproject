```python
from pycaret.datasets import get_data
import pandas as pd
from pycaret.regression import *
```

```python
all_datasets = get_data('index')
#Para checar los nombres de los datasets disponibles,
#Se encontro el dataset 'house' que es el de house prices de kaggle
data_house = get_data(dataset='house')
```

EDA y profiling

```python
data_house = get_data(dataset='house', profile=True)
```

Warnings

Las variables que tienen warnings de muchos datos faltantes son: 'Alley', 'PoolQC', 'MiscFeature', 'Fence', 'FireplaceQu', 'Utilities'

Estas se eliminarn del dataset

La imputacin de datos ser por medio de eliminacin ya que no hay porcentajes muy altos

Hay peligro de multicolinealidad por lo que se activar remove_multicollinearity

```python
num_filas_train = int(len(data_house)*0.8)

data_train = data_house.sample(n=num_filas_train, random_state=2023)
data_test = data_house.drop(data_train.index)
```

```python
print(f"Filas en Train: {data_train.shape[0]}")
print(f"Filas en Test: {data_test.shape[0]}")
```

```python
columns_to_drop = ['Alley','PoolQC','MiscFeature','Fence','FireplaceQu','Utilities']
data_train = data_train.drop(columns=columns_to_drop)
data_test = data_test.drop(columns=columns_to_drop)
```

Setup del modelo

```python
dataset = setup(data = data_train,
                target = 'SalePrice',
                session_id=2023,
                normalize=True,
                normalize_method='minmax',
                transformation=True,
                transformation_method = 'yeo-johnson',
                transform_target = True,
                log_experiment=True,
                remove_multicollinearity = True,
                experiment_name='Sale Price predict'
                )
```

Entrenamiento y seleccion de modelos

```python
best = compare_models(sort='RMSE')
```

```python
print(best)
```

```python
tuned_best=tune_model(estimator=best,optimize='RMSE')
```

Eleccion de modelo

Top 3 modelos

Gradient Boosting Regressor; RMSE: 28081.2867; R2:	0.8462

Light Gradient Boosting Machine; RMSE: 28710.0123 R2:	0.8471

Extreme Gradient Boosting; RMSE: 29348.6900; R2:	0.8459

```python
#Se elige el modelo Gradien Boosting Regressor por tener el mejor RMSE y a pesar de no tener el mejor R2, tiene mejor relacion
final_model = finalize_model(estimator=best)
```
